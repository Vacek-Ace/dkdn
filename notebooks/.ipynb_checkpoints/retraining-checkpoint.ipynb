```python
summary = pd.concat([summary_retraining(path=f'../results/incremental/conf{i+1}', setting_col=True) for i in range(6)] , ignore_index=True)
```

```python
df_plot = summary.groupby(['setting', 'model', 'proportion'])[['performance gap', 'sample']].mean().reset_index()
```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
for model in df_plot['model'].unique():
    sns_plot = sns.catplot(x="proportion", y="performance gap", hue="setting", #row='model',
                    capsize=.2, palette="husl", height=8, aspect=1.5,sharey=True,
                    kind="point", data=df_plot[df_plot['model'] == model], legend=True, legend_out=False)
    sns_plot.despine(left=True).set_axis_labels(f"{model}")
```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
for model in df_plot['model'].unique():
    sns_plot = sns.catplot(x="proportion", y="sample", hue="setting", #row='model',
                    capsize=.2, palette="husl", height=8, aspect=1.5,sharey=True,
                    kind="point", data=df_plot[df_plot['model'] == model], legend=True, legend_out=False)
    sns_plot.despine(left=True).set_axis_labels(f"{model}")
```
